```python
import os
import mne
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print(__doc__)
```

```python
def fetch(id, col):
    print('-' * 80)
    print(id)
    path = fifs[fifs['id'] == id][col].values[0]
    print(path)
    return path
fifs = pd.read_json('fifs.json')
fifs
```

```python
tmin = -0.2
picks = ('meg', 'mag')
raw = mne.concatenate_raws([mne.io.read_raw_fif(fetch('ZYF_Test_{}'.format(j), col='rawica')) for j in [3, 5, 7]])
events = mne.find_events(raw)
epochs = mne.Epochs(raw, events, picks=picks, tmin=tmin, tmax=3, decim=10)
event_id = epochs.event_id
```

```python
epochs
```

```python
xdawn = mne.preprocessing.Xdawn(n_components=6)
xdawn.fit(epochs)
```

```python
epochs.load_data()
e = xdawn.apply(epochs)
```

```python
e
```

```python
epo = e['8']
epo.apply_baseline(baseline=(tmin, 0))
epo.plot_psd(spatial_colors=True, average=True)
for event in event_id:
    print(event)
    evoked = epo[event].average()
    evoked.plot_joint(ts_args=dict(gfp=True))
```
