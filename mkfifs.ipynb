```python
import os
import pandas as pd
```

```python
fifs = pd.DataFrame(columns=('id', 'rawpath', 'rawica'))
for subject in ['QYJ', 'ZYF']:
    for task, jj in [('Test', 8), ('Training', 5)]:
        for j in range(0+1, jj+1):
            fifs = fifs.append(
                [{
                    'id': '{subject}_{task}_{j}'.format(subject=subject, task=task, j=j),
                    'rawpath': os.path.join('rawdata', subject, 'Multi{task}_{j}_raw_tsss.fif'.format(task=task, j=j)),
                    'rawica': os.path.join('raw_ica', '{subject}_{task}_{j}-afterica-raw.fif'.format(subject=subject, task=task, j=j))
                }],
                ignore_index=True)

fifs.to_json('fifs.json')
fifs
```
